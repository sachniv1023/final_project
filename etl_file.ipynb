## Creating tables for our loan database

```python
import psycopg2


# Connecting to database 
conn = psycopg2.connect(
           host="localhost",
           dbname= "loan",
           user="postgres",
           password="nivedita1023"
)
cur = conn.cursor()
```

```python
cur.execute("""CREATE TABLE education(
    education_id INTEGER Primary Key,
    education_level VARCHAR 
)
""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE income(
    income_id INTEGER Primary Key,
    income INTEGER 
)
""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE mortgage(
    mortgage_id INTEGER Primary Key,
    mortgage INTEGER 
)
""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE customer_details(
    cust_id INTEGER PRIMARY KEY,
    age INTEGER,
    experience INTEGER,
    zipcode INTEGER,
    family INTEGER,
    personal_loan VARCHAR,
    creditcard VARCHAR,
    mortgage_id INTEGER REFERENCES mortgage,
    education_id INTEGER REFERENCES education,
    education_level VARCHAR,
    securities_account VARCHAR,
    cd_account VARCHAR,
    ccavg FLOAT
    
)
""")
conn.commit()
```

```python
cur.execute("""CREATE TABLE customer_income(
    income_id INTEGER REFERENCES income,
    cust_id INTEGER REFERENCES customer_details
)
""")
conn.commit()
```

### Reading Data From Database using (read_sql)

#### Here we are reading data from our tables(customer_details, educatiion, income, customer_income, mortgage) to a pandas DataFrame.We call pd.read_sql() for achieving this and passing the database connection obtained from the SQLAlchemy Engine as a parameter.

Importing libraries

```python
##importing python library psycopg2 inorder to work with databases.
import psycopg2

#importing pandas
import pandas as pd

#To read from and write to SQL end-points like PostgreSQL the pandas DataFrame class uses the SQLAlchemy engine.
from sqlalchemy import create_engine

#Create an engine instance

alchemyEngine  = create_engine('postgresql+psycopg2://postgres:nivedita1023@localhost:5432/loan') # this is our connection string
```

```python
#Connect to PostgreSQL server

dbConnection = alchemyEngine.connect();
```

```python
#Read data from PostgreSQL database table and load into a DataFrame instance


dataFrame  = pd.read_sql_table("education", dbConnection);
print(dataFrame)

# Close the database connection

#dbConnection.close();
```

```python
#Read data from PostgreSQL database table and load into a DataFrame instance


dataFrame  = pd.read_sql_table("customer_details", dbConnection);
print(dataFrame)

# Close the database connection

#dbConnection.close();
```

### Query Execution

```python
#find out the customer who's age is between 30 to 60 and their experience and education_level
query_1 = '''
select cust_id,age,experience,education_level
from customer_details
WHERE age >30
order by age desc
'''
##Read SQL query into a DataFrame
df1 = pd.read_sql_query(query_1,dbConnection)
df1
```

```python
#find out customer who's income is more than 180.
query_2 = '''
select  age, education_level, income
from customer_details
inner join customer_income
on customer_details.cust_id=customer_income.cust_id
inner join income
on income.income_id=customer_income.income_id
where income>180;
'''
##Read SQL query into a DataFrame
df2 = pd.read_sql_query(query_2,dbConnection)
df2
```

```python
#find out customer who r interedsted in personal loan and having family size=3, and higher educated. 
query_3 = '''
Select education_level,family,personal_loan
from customer_details 
where
family = 3 and education_level= 'advance' and personal_loan= 'Yes';
'''
##Read SQL query into a DataFrame
df3 = pd.read_sql_query(query_3,dbConnection)
df3
```

```python
#information about customer age, education level and income is greater than 100.
query_4 = '''
select  age, education_level, income
from customer_details
inner join customer_income
on customer_details.cust_id=customer_income.cust_id
inner join income
on income.income_id=customer_income.income_id
where income>100;
'''
##Read SQL query into a DataFrame
df4 = pd.read_sql_query(query_4,dbConnection)
df4.head()
```

```python
#find out the customer who's mortgage is greater than 0 and ccavg is greater than 1.
query_5 = '''
select  age, education_level,cust_id, family, ccavg, mortgage
from customer_details
inner join mortgage
on customer_details.mortgage_id=mortgage.mortgage_id
where mortgage>'0' and ccavg>1 ;
'''
##Read SQL query into a DataFrame
df5 = pd.read_sql_query(query_5,dbConnection)
df5
```
